# Hyperbolic Mohr-Coulomb criterion

__*Load some Python modules*__

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
from math import *
from collections import OrderedDict
```

__*Update some rcParams*__

```python
plt.rcParams.update({
    "text.usetex":True,
    "font.family":"serif",
    "text.latex.preamble":r"\usepackage[T1]{fontenc}\usepackage[utf8]{inputenc}\usepackage{amsmath}\usepackage{amssymb}\usepackage{amsfonts}\usepackage{fourier}",
    "font.size":18,
    'figure.constrained_layout.use':True,
    'figure.constrained_layout.h_pad':.25,
    'figure.constrained_layout.w_pad':.25,
})
```

__*Define the color map for the figures*__

```python
MyCMap=plt.get_cmap("jet")
MyCMapUP=MyCMap(np.arange(MyCMap.N))
```

__*Define a dictionary with the default values of the material parameters*__

```python
DefaultDicPara=OrderedDict()
DefaultDicPara["FrictionAngle"]=30.
DefaultDicPara["DilationAngle"]=30.
DefaultDicPara["Cohesion"]=1
DefaultDicPara["TensionCutOff"]=.2*DefaultDicPara["Cohesion"]/tan(DefaultDicPara["FrictionAngle"]*pi/180.)
DefaultDicPara["PiShapeCoef"]=.9995
```

__*Define a function for the hyperbolic Mohr-Coulomb plastic criterion*__
$$F(\boldsymbol{X})\mathrel{\mathop:}= X_m\sin{\varphi}+\sqrt{\frac{X_{eq}^2}{3}{K(\theta)}^2+a^2\sin^2{\varphi}}-C\cos{\varphi},\quad X_m=\frac{X_{kk}}{3},\quad X_{eq}=\sqrt{\frac{3}{2}\boldsymbol{X_d}:\boldsymbol{X_d}}$$
__*With*__
$$K(\theta)=K_c\cos\left(\frac{\beta\pi}{6}-\frac{\cos^{-1}\left(\gamma\cos{3\theta}\right)}{3}\right),\quad \beta=\frac{6}{\pi}\tan^{-1}\left(\frac{2r-1}{\sqrt{3}}\right),\quad r=\frac{3-\sin{\varphi}}{3+\sin{\varphi}},\quad K_c=\frac{\sqrt{3}}{6}\frac{3-\sin{\varphi}}{\cos\left(\frac{\beta\pi}{6}-\frac{\cos^{-1}{\left(-\gamma\right)}}{3}\right)},\quad \gamma\in\left[0,1\right[$$

```python
def Fp(Xeq,Xm,Theta,DPara):
    # >>> Get the material parameters
    phi=DPara["FrictionAngle"]*pi/180.
    c=DPara["Cohesion"]
    a=DPara["TensionCutOff"]
    gamma=DPara["PiShapeCoef"]
    # >>> Compute the function K(t)
    r=(3.-sin(phi))/(3.+sin(phi))
    beta=6./pi*atan((2.*r-1.)/sqrt(3.))
    Kc=sqrt(3.)/6.*(3.-sin(phi))/cos(beta*pi/6.-acos(-gamma)/3.)
    K=Kc*cos(beta*pi/6.-acos(gamma*cos(3.*Theta))/3.)
    # >>> Output
    return Xm*sin(phi)+sqrt(Xeq*Xeq*K*K/3.+a*a*sin(phi)*sin(phi))-c*cos(phi)
```

__*Define a function for the hyperbolic Mohr-Coulomb plastic flow function*__
$$G(\boldsymbol{X})\mathrel{\mathop:}= X_m\sin{\psi}+\sqrt{\frac{X_{eq}^2}{3}{K_G(\theta)}^2+a^2\tan^2{\varphi}\cos^2{\psi}}-C\cos{\psi},\quad X_m=\frac{X_{kk}}{3},\quad X_{eq}=\sqrt{\frac{3}{2}\boldsymbol{X_d}:\boldsymbol{X_d}}$$
__*With*__
$$K_G(\theta)=K_G^c\cos\left(\frac{\beta_G\pi}{6}-\frac{\cos^{-1}\left(\gamma\cos{3\theta}\right)}{3}\right),\quad \beta_G=\frac{6}{\pi}\tan^{-1}\left(\frac{2r_G-1}{\sqrt{3}}\right),\quad r_G=\frac{3-\sin{\psi}}{3+\sin{\psi}},\quad K_G^c=\frac{\sqrt{3}}{6}\frac{3-\sin{\psi}}{\cos\left(\frac{\beta_G\pi}{6}-\frac{\cos^{-1}{\left(-\gamma\right)}}{3}\right)},\quad \gamma\in\left[0,1\right[$$

```python
def Gp(Xeq,Xm,Theta,DPara):
    # >>> Get the material parameters
    phi=DPara["FrictionAngle"]*pi/180.
    psi=DPara["DilationAngle"]*pi/180.
    c=DPara["Cohesion"]
    a=DPara["TensionCutOff"]
    gamma=DPara["PiShapeCoef"]
    # >>> Compute the function K(t)
    rG=(3.-sin(psi))/(3.+sin(psi))
    betaG=6./pi*atan((2.*rG-1.)/sqrt(3.))
    KGc=sqrt(3.)/6.*(3.-sin(psi))/cos(betaG*pi/6.-acos(-gamma)/3.)
    KG=KGc*cos(betaG*pi/6.-acos(gamma*cos(3.*Theta))/3.)
    # >>> Output
    return Xm*sin(psi)+sqrt(Xeq*Xeq*KG*KG/3.+a*a*tan(phi)*tan(phi)*cos(psi)*cos(psi))-c*cos(psi)
```

__*Define a dictionary for a parametric study on the material parameters*__

```python
MCTension=DefaultDicPara["Cohesion"]/tan(DefaultDicPara["FrictionAngle"]*pi/180.)
DSensi=OrderedDict()
DSensi["PiShapeCoef"]={"list":np.array([0.,.1,.3,.6,.8,.9,.99,1.]),"label":r"$\gamma$ (-)","scale":1.}
DSensi["Cohesion"]={"list":np.linspace(.5,10.,8),"label":r"$C$ (MPa)","scale":1.}
DSensi["FrictionAngle"]={"list":np.linspace(5.,40.,8),"label":r"$\varphi$ (°)","scale":1.}
DSensi["DilationAngle"]={"list":np.linspace(5.,40.,8),"label":r"$\psi$ (°)","scale":1.}
DSensi["TensionCutOff"]={"list":np.linspace(.1,.9,8)*MCTension,"label":r"$a$ (MPa)","scale":1.}
```

__*Create plots of $F^p$ in the $\pi$ plane for $X_m\equiv 0$*__

```python
### >>> Array with the values for Theta
lTheta=np.linspace(0.,2.*pi,599)

### >>> Define the figure and the subplots
fig,ax=plt.subplots(2,2,subplot_kw={'projection':'polar'},figsize=(14,15))
fx=ax.flatten()

### >>> Loop on DSensi items
i=0
for p,dic in DSensi.items():
    if p!="DilationAngle":
        ### >>> Temporary dictionary for le loop
        DTmp=DefaultDicPara.copy()
        ### >>> Loop on the values of the parameter p
        j=0
        for vp in dic["list"]:
            lXeq=[]
            DTmp[p]=vp
            ### >>> Loop on Theta
            for t in lTheta:
                root=fsolve(Fp,1.,args=(0.,t,DTmp))
                lXeq.append(root[0])
            ### >>> Plot
            ic=float(j)/max(1.,float(len(dic["list"])-1.))
            ic=int(ic*(len(MyCMapUP)-1.))
            fx[i].plot(lTheta,lXeq,lw=2,c=MyCMapUP[ic])
            j+=1
        ### >>> Options for subplots axis
        fx[i].set_theta_offset(pi/6.)
        fx[i].set_xticks(np.linspace(0.,5.*pi/3.,6))
        fx[i].set_xticklabels([r"$0$",r"$\frac{\pi}{3}$",r"$\frac{2\pi}{3}$",r"$\pi$",r"$\frac{4\pi}{3}$",r"$\frac{5\pi}{3}$"])
        fx[i].set_rticks([])
        fx[i].set_facecolor('.7')
        fx[i].grid(lw=1,ls=':',c='.5',zorder=0)
        ### >>> Options for color bars
        norm=mpl.colors.Normalize(vmin=0.,vmax=1.)
        cb=fig.colorbar(mpl.cm.ScalarMappable(norm=norm,cmap=MyCMap),ax=fx[i],orientation='horizontal',shrink=.9,aspect=35,location="top")
        cb.set_ticks(np.linspace(0.,1.,len(dic["list"])))
        cb.set_ticklabels(['{level:.2g}'.format(level=l/dic["scale"]) for l in dic["list"]])
        t=cb.set_label(dic["label"],va="bottom")
        i+=1

### >>> Options for the figure
fig.set_facecolor('.9')

### >>> Save the figure
plt.savefig("hyperbolic_mc_fp_pi_plane_comp.pdf",dpi=600,bbox_inches='tight')
```

__*Create plots of $F^p$ in the stress plane ${\left(X_m,X_{eq}\right)}$ for $\theta=0$ (extension)*__

```python
### >>> Define the figure and the subplots
fig,ax=plt.subplots(2,2,figsize=(16,12))
fx=ax.flatten()

### >>> Loop on DSensi items
i=0
for p,dic in DSensi.items():
    if p!="DilationAngle":
        ### >>> Temporary dictionary for le loop
        DTmp=DefaultDicPara.copy()
        ### >>> Loop on the values of the parameter p
        j=0
        for vp in dic["list"]:
            lXeq=[]
            DTmp[p]=vp
            MCT=DTmp["Cohesion"]/tan(DTmp["FrictionAngle"]*pi/180.)-DTmp["TensionCutOff"]
            if p=="TensionCutOff":
                lXm=np.linspace(-5.,.9999*MCT,1000)
            else:
                lXm=np.linspace(-15.,.9999*MCT,1000)
            ### >>> Loop on Xm
            for Xm in lXm:
                root=fsolve(Fp,1.,args=(Xm,0.,DTmp))
                lXeq.append(root[0])
            ### >>> Plot
            ic=float(j)/max(1.,float(len(dic["list"])-1.))
            ic=int(ic*(len(MyCMapUP)-1.))
            fx[i].plot(lXm,lXeq,lw=2,c=MyCMapUP[ic],zorder=10)
            j+=1
        ### >>> Options for subplots axis
        fx[i].set_xlabel(r"$X_m$ (MPa)")
        fx[i].set_ylabel(r"$X_{eq}$ (MPa)")
        fx[i].minorticks_on()
        fx[i].set_facecolor('.7')
        fx[i].grid(lw=1,ls=':',c='.5',zorder=0)
        fx[i].axhline(0.,lw=1,ls='-',c='.4',zorder=3)
        fx[i].axvline(0.,lw=1,ls='-',c='.4',zorder=3)
        ### >>> Options for color bars
        norm=mpl.colors.Normalize(vmin=0.,vmax=1.)
        cb=fig.colorbar(mpl.cm.ScalarMappable(norm=norm,cmap=MyCMap),ax=fx[i],orientation='horizontal',shrink=.9,aspect=35,location="top")
        cb.set_ticks(np.linspace(0.,1.,len(dic["list"])))
        cb.set_ticklabels(['{level:.2g}'.format(level=l/dic["scale"]) for l in dic["list"]])
        t=cb.set_label(dic["label"],va="bottom")
        i+=1

### >>> Options for the figure
fig.set_facecolor('.9')

### >>> Save the figure
plt.savefig("hyperbolic_mc_fp_stress_plane_comp.pdf",dpi=600,bbox_inches='tight')
```

__*Create plots of $G^p$ in the $\pi$ plane for $X_m\equiv 0$*__

```python
### >>> Array with the values for Theta
lTheta=np.linspace(0.,2.*pi,600)

### >>> Define the figure and the subplots
fig,ax=plt.subplots(2,3,subplot_kw={'projection':'polar'},figsize=(20,15))
fx=ax.flatten()

### >>> Loop on DSensi items
i=0
for p,dic in DSensi.items():
    ### >>> Temporary dictionary for le loop
    DTmp=DefaultDicPara.copy()
    ### >>> Loop on the values of the parameter p
    j=0
    for vp in dic["list"]:
        lXeq=[]
        DTmp[p]=vp
        ### >>> Loop on Theta
        for t in lTheta:
            root=fsolve(Gp,1.,args=(0.,t,DTmp))
            lXeq.append(root[0])
        ### >>> Plot
        ic=float(j)/max(1.,float(len(dic["list"])-1.))
        ic=int(ic*(len(MyCMapUP)-1.))
        fx[i].plot(lTheta,lXeq,lw=2,c=MyCMapUP[ic])
        j+=1
    ### >>> Options for subplots axis
    fx[i].set_theta_offset(pi/6.)
    fx[i].set_xticks(np.linspace(0.,5.*pi/3.,6))
    fx[i].set_xticklabels([r"$0$",r"$\frac{\pi}{3}$",r"$\frac{2\pi}{3}$",r"$\pi$",r"$\frac{4\pi}{3}$",r"$\frac{5\pi}{3}$"])
    fx[i].set_rticks([])
    fx[i].set_facecolor('.7')
    fx[i].grid(lw=1,ls=':',c='.5',zorder=0)
    ### >>> Options for color bars
    norm=mpl.colors.Normalize(vmin=0.,vmax=1.)
    cb=fig.colorbar(mpl.cm.ScalarMappable(norm=norm,cmap=MyCMap),ax=fx[i],orientation='horizontal',shrink=.9,aspect=35,location="top")
    cb.set_ticks(np.linspace(0.,1.,len(dic["list"])))
    cb.set_ticklabels(['{level:.2g}'.format(level=l/dic["scale"]) for l in dic["list"]])
    t=cb.set_label(dic["label"],va="bottom")
    i+=1

### >>> Options for the figure
fig.set_facecolor('.9')
fx[-1].set_visible(False)

### >>> Save the figure
plt.savefig("hyperbolic_mc_gp_pi_plane_comp.pdf",dpi=600,bbox_inches='tight')
```

__*Create plots of $G^p$ in the stress plane ${\left(X_m,X_{eq}\right)}$ for $\theta=0$ (extension)*__

```python
### >>> Define the figure and the subplots
fig,ax=plt.subplots(2,3,figsize=(20,12))
fx=ax.flatten()

### >>> Loop on DSensi items
i=0
for p,dic in DSensi.items():
    ### >>> Temporary dictionary for le loop
    DTmp=DefaultDicPara.copy()
    ### >>> Loop on the values of the parameter p
    j=0
    for vp in dic["list"]:
        lXeq=[]
        DTmp[p]=vp
        MCT=DTmp["Cohesion"]/tan(DTmp["DilationAngle"]*pi/180.)-DTmp["TensionCutOff"]*tan(DTmp["FrictionAngle"]*pi/180.)/tan(DTmp["DilationAngle"]*pi/180.)
        if p=="TensionCutOff":
            lXm=np.linspace(-5.,.9999*MCT,1000)
        else:
            lXm=np.linspace(-15.,.9999*MCT,1000)
        ### >>> Loop on Theta
        for Xm in lXm:
            root=fsolve(Gp,1.,args=(Xm,0,DTmp))
            lXeq.append(root[0])
        ### >>> Plot
        ic=float(j)/max(1.,float(len(dic["list"])-1.))
        ic=int(ic*(len(MyCMapUP)-1.))
        fx[i].plot(lXm,lXeq,lw=2,c=MyCMapUP[ic],zorder=10)
        j+=1
    ### >>> Options for subplots axis
    fx[i].set_xlabel(r"$X_m$ (MPa)")
    fx[i].set_ylabel(r"$X_{eq}$ (MPa)")
    fx[i].minorticks_on()
    fx[i].set_facecolor('.7')
    fx[i].grid(lw=1,ls=':',c='.5',zorder=0)
    fx[i].axhline(0.,lw=1,ls='-',c='.4',zorder=3)
    fx[i].axvline(0.,lw=1,ls='-',c='.4',zorder=3)
    ### >>> Options for color bars
    norm=mpl.colors.Normalize(vmin=0.,vmax=1.)
    cb=fig.colorbar(mpl.cm.ScalarMappable(norm=norm,cmap=MyCMap),ax=fx[i],orientation='horizontal',shrink=.9,aspect=35,location="top")
    cb.set_ticks(np.linspace(0.,1.,len(dic["list"])))
    cb.set_ticklabels(['{level:.2g}'.format(level=l/dic["scale"]) for l in dic["list"]])
    t=cb.set_label(dic["label"],va="bottom")
    i+=1

### >>> Options for the figure
fig.set_facecolor('.9')
fx[-1].set_visible(False)

### >>> Save the figure
plt.savefig("hyperbolic_mc_gp_stress_plane_comp.pdf",dpi=600,bbox_inches='tight')
```
